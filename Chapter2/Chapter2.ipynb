```python
%matplotlib notebook
import math
import numpy as np
import matplotlib.pyplot as plt
N = 5
a = [(2*math.pi*math.sin(2*math.pi*n) + math.cos(2*math.pi*n) - 1)/(2*(math.pi**2)*n**2) for n in range(1,N+1)]
b = [(math.sin(2*math.pi*n) - 2*math.pi*n*math.cos(2*math.pi*n))/(2*(math.pi**2)*n**2) for n in range(1,N+1)]

def fourier(x):
    a_0 = 1
    y = a_0/2 + sum([a[n-1]*math.cos(2*math.pi*n*x) + b[n-1]*math.sin(2*math.pi*n*x) for n in range(1,N+1)])
    return y


X = np.linspace(0,1,1000)
Y = [fourier(x) for x in X]
plt.scatter(X,Y)
plt.show()
```
